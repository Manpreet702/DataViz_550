The original scatterplot conveys very little information about how satisfaction has changed because it omits important context.  It uses the default ggplot scales, so the axes extend beyond the range of proportions (0–1), and the reader has no way to see that the data are proportions.  Without a reference line it is hard to judge whether points lie above or below the 45° line.  The plot title “Dramatic increase in dissatisfaction at work” is not supported by the data: the majority of cities lie close to the diagonal, indicating that the proportion of dissatisfied workers has changed only slightly from last year to this year, and in many places it has actually decreased.  A more appropriate interpretation is that job dissatisfaction levels vary widely between cities, but on average have remained relatively stable between the two years; there is no evidence of a dramatic increase.

```r
# load packages
library(ggplot2)
library(readr)
library(dplyr)
library(scales)

# read the data
jobs <- read_csv("cities-job-satisfaction.csv")

# compute limits
max_prop <- max(c(jobs$last_year, jobs$this_year), na.rm = TRUE)

# an improved plot: show identity line and proper limits
ggplot(jobs, aes(x = last_year, y = this_year)) +
  geom_abline(intercept = 0, slope = 1, linetype = "dashed", colour = "red") +
  geom_point(alpha = 0.5, colour = "steelblue") +
  scale_x_continuous(
    limits = c(0, max_prop),
    labels = percent_format(accuracy = 1),
    name = "Proportion dissatisfied last year"
  ) +
  scale_y_continuous(
    limits = c(0, max_prop),
    labels = percent_format(accuracy = 1),
    name = "Proportion dissatisfied this year"
  ) +
  labs(
    title = "Work dissatisfaction this year vs last year by city",
    subtitle = "Each point represents one city; the dashed line indicates no change"
  ) +
  theme_minimal(base_size = 12)
```

The bar chart in the mayor's plot counts how many companies were surveyed before and after the election.  These counts are identical (both groups have 61 companies), but the number of observations tells us nothing about salaries or changes in wages.  A bar chart of counts therefore cannot be used to support a claim about pay increases.

In the other bar charts the mayor summarises the data with a single number (either the mean or the median) for each group and then compares those numbers as if they were paired measurements.  However, the data are not paired: each company contributes one salary figure to the “before” group and a different set of companies may contribute to the “after” group.  Furthermore, the distributions are skewed, with several very high or low salaries, so differences in the mean or median alone do not tell the whole story.  A visualisation of the full distribution shows that while the typical salary after the election is somewhat higher, there is still a wide spread, and some companies even report lower average salaries.  Thus it is not accurate to claim that wages have been raised for all companies.

```r
# load packages
library(ggplot2)
library(readr)
library(dplyr)

# read the data
wages <- read_csv("wages.csv")

# count number of observations in each group
count_df <- wages %>% count(when)

# recreate the mayor's count bar chart
ggplot(count_df, aes(y = when, x = n, fill = when)) +
  geom_col() +
  geom_text(aes(label = n), hjust = -0.1, colour = "black") +
  scale_x_continuous(expand = expansion(mult = c(0, 0.05))) +
  labs(
    title = "Number of companies surveyed",
    x = "Count",
    y = NULL
  ) +
  scale_fill_manual(values = c("after" = "#F8766D", "before" = "#00BFC4"), guide = "none") +
  theme_minimal()
```

```r
# calculate mean salary for each group
mean_df <- wages %>%
  group_by(when) %>%
  summarise(mean_salary = mean(salary, na.rm = TRUE))

# recreate the mayor's mean salary bar chart
ggplot(mean_df, aes(y = when, x = mean_salary, fill = when)) +
  geom_col() +
  geom_text(aes(label = sprintf("$%.2f", mean_salary)), hjust = -0.1, colour = "black") +
  scale_x_continuous(expand = expansion(mult = c(0, 0.05)), labels = scales::dollar_format(prefix = "$")) +
  labs(
    title = "Mean company salary",
    x = "Average salary",
    y = NULL
  ) +
  scale_fill_manual(values = c("after" = "#F8766D", "before" = "#00BFC4"), guide = "none") +
  theme_minimal()
```

```r
# calculate median salary for each group
median_df <- wages %>%
  group_by(when) %>%
  summarise(median_salary = median(salary, na.rm = TRUE))

# recreate the mayor's median salary bar chart
ggplot(median_df, aes(y = when, x = median_salary, fill = when)) +
  geom_col() +
  geom_text(aes(label = sprintf("$%.2f", median_salary)), hjust = -0.1, colour = "black") +
  scale_x_continuous(expand = expansion(mult = c(0, 0.05)), labels = scales::dollar_format(prefix = "$")) +
  labs(
    title = "Median company salary",
    x = "Median salary",
    y = NULL
  ) +
  scale_fill_manual(values = c("after" = "#F8766D", "before" = "#00BFC4"), guide = "none") +
  theme_minimal()
```

The mayor's plot is highly misleading.  First, the vertical axis is reversed so that higher numbers of firearm murders appear lower on the page.  This inversion makes it look as though murders “plummet” after the gun law when in fact the values are simply plotted upside down.  Second, the shaded area and the red annotation give the impression of a dramatic change at the time of the law, yet a cursory look at the data shows that the decline in firearm murders started long before 2005 and that there was actually an uptick after 2010.  Finally, the plot lacks proper context: the y‐axis does not start at zero, and no uncertainty or comparison to other states is provided.  A more honest interpretation is that firearm murders fluctuated over time, with a decreasing trend through the 1990s, a modest rise in the 2000s, and a sharper increase after 2010.  There is no clear evidence that the 2005 law caused a sudden and sustained drop in deaths.

The density plot presented by the mayor is problematic for several reasons.  Approval ratings are discrete integers between 0 and 10 000, yet a continuous kernel density estimate is used; this smooths away important features and suggests values that are impossible (e.g., negative ratings).  The title “Ocean waves” and the unlabeled densities are uninformative and do not help the reader understand what is being plotted.  Labelling the tick marks as “2K”, “5K”, “8K” and “10K” but leaving off the units on the axis obscures the magnitude of the ratings.  Most importantly, overlaying the two densities makes it difficult to compare distributions: one cannot easily see where one group has higher or lower approval, and the curves imply differences that may not be statistically significant.  A more appropriate summary would show the full distribution or key summary statistics side by side, clearly indicating that the two states have similar spreads and that the neighbour state tends to have slightly higher approval ratings overall.
