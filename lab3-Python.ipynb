### Mistakes in the Plot and Claim

* **Misleading Transparency Claim**: The claim suggests that the dark area represents an even distribution of cities across the 0% to 40% dissatisfaction range. However, even with transparency, the high density of points near the bottom (0% dissatisfaction) creates a very dark cluster, indicating that the vast majority of cities actually stayed near 0%.
* **Overplotting Issues**: The scatter plot format makes it difficult to see the true concentration of data points. While transparency is used, it does not sufficiently reveal that the "dramatic increase" applies only to a few outlier cities, not the general population of cities.
* **Inaccurate Generalization**: The claim that "people are dramatically more dissatisfied" is a generalization based on a subset of the data. Most cities are clustered at the bottom left, meaning dissatisfaction remained low and stable for most locations.

### Correct Interpretation of the Data

* **Stability for Most Cities**: An investigation of the raw data in `cities-job-satisfaction.csv` shows that for the majority of cities, the proportion of dissatisfied workers remained consistently low (near 0%) for both last year and this year.
* **Outliers vs. Trend**: There is no "even amount of cities" increasing to 40% dissatisfaction. Instead, there are a few significant outliers where dissatisfaction spiked, but these do not represent a global or "dramatic" general trend.
* **Improved Visualization Choice**: A better way to represent this would be a **histogram** of the differences (this year minus last year) or a **heatmap/binning plot**, which would more clearly show that the bulk of the data is concentrated near the origin (no change/low dissatisfaction).

```python
# YOUR ANSWER HERE
import altair as alt
import pandas as pd
alt.data_transformers.enable('default')
# Load the data
df_cities = pd.read_csv('.//data//cities-job-satisfaction.csv')

# Recreate the campaign plot
q1_2_plot = alt.Chart(df_cities).mark_circle(
    opacity=0.3,
    color='#222222',
    size=60
).encode(
    alt.X('last_year:Q', 
          title='Proportion of dissatisfied workers last year',
          scale=alt.Scale(domain=[0, 0.6])),
    alt.Y('this_year:Q', 
          title='Proportion of dissatisfied workers this year',
          scale=alt.Scale(domain=[0, 0.6]))
).properties(
    title='Dramatic decline in job satisfaction',
    width=400,
    height=400
).configure_axis(
    grid=True
).configure_view(
    strokeWidth=0
)

q1_2_plot
```

#### Mistakes in the Claim and Plots:
The "Flaw of Averages": The claim states that because the average and median increased, "almost all workers" are receiving higher pay. This is a classic logical fallacy. While the center of the distribution moved up, the spread (variance) increased drastically.
Hidden Inequality: By only showing summary statistics (mean and median), the plots hide the fact that a large portion of the population is actually worse off.
Interpretation of the Data: Upon inspecting the wages.csv data, we see that before the election, salaries were tightly clustered between $\$15$ and $\$25$. After the election, the distribution became bimodal or highly skewed. In the "after" group, the bottom 25% of workers earn around $\$12.50$, which is lower than the absolute minimum salary before $(\$15.20)$.

```python
# YOUR ANSWER HERE
import altair as alt
import pandas as pd

wages = pd.read_csv('./data//wages.csv')

# Plot 1: Mean Salary
plot_mean = alt.Chart(wages).mark_bar().encode(
    alt.X('when:N', title=None),
    alt.Y('mean(salary):Q', title='Average Salary')
).properties(width=200, height=300, title="Average Salary")

# Plot 2: Median Salary
plot_median = alt.Chart(wages).mark_bar().encode(
    alt.X('when:N', title=None),
    alt.Y('median(salary):Q', title='Median Salary')
).properties(width=200, height=300, title="Median Salary")

# Plot 3: Sample Size (Count)
plot_count = alt.Chart(wages).mark_bar().encode(
    alt.X('when:N', title=None),
    alt.Y('count():Q', title='Number of workers')
).properties(width=200, height=300, title="Number of workers")

# Displaying them (usually shown separately or side-by-side)
plot_mean
```

#### Mistakes in the Plot and Claim:
Inverted Y-Axis: This visualization is a famous example of deceptive design. The Y-axis has been inverted, placing $0$ at the top and the maximum value at the bottom. This tricks the eye into thinking a line moving "downward" represents a decrease, when it actually represents a significant increase in the number of murders.Deceptive Shading: The use of red shading above the line (which is actually the area representing lower values due to the inversion) creates a "dripping" effect that misleads the viewer into focusing on the white space below as the "trend."Factual Misrepresentation: The claim states that deaths "plummeted" after the law was enacted. By looking at the actual numbers on the axis, it is clear that the opposite is true.
#### Correct Interpretation of the Data:Post-Law Surge: 
According to the guns-r-us.csv data, firearm murders in Florida were at a relative low of approximately 234 in 2006 (the year after the law). Following this, there was a sharp and steady increase.Peak Values: By 2011, the number of firearm murders had risen to over 700, nearly tripling the 2006 figures.Conclusion: The enactment of the law was followed by a dramatic upward trend in firearm-related deaths, directly contradicting the visual suggestion of the inverted chart.

```python

# Load the data
df_guns = pd.read_csv('./data//guns-r-us.csv')

# Recreate the campaign plot exactly
# We combine an area mark with a point mark to match your request
base = alt.Chart(df_guns).encode(
    alt.X('Year:O', title='Year'),
    alt.Y('Firearm murders:Q', 
          title='Number of murders by firearms',
          scale=alt.Scale(reverse=True, domain=[0, 900])) # Inverted and scaled
)

# The red area "dripping" from the top
area = base.mark_area(
    color='#C0392B', # Darker campaign red
    opacity=0.8,
    interpolate='monotone'
)

# Adding the points as requested
points = base.mark_point(
    color='#000000',
    filled=True,
    size=50
)

q3_2_plot = (area + points).properties(
    title='Gun deaths in Florida',
    width=600,
    height=400
).configure_view(
    strokeWidth=0
).configure_axis(
    grid=True
)

q3_2_plot
```

### Question 3.3
rubric={viz:2}

<h4>Python</h4>
<p>Improve my Altair plot. Chose the best type of plot to accurately reveal what the data looks like. Design the plot to be an effective plot for communication purposes, this includes proper figure size, axis/figure titles, font sizes, colors, etc.</p>
    
</div>

```python
# YOUR ANSWER HERE
# Load the data
df_guns = pd.read_csv('./data//guns-r-us.csv')

# 1. Create the standard line chart
base = alt.Chart(df_guns).encode(
    alt.X('Year:O', title='Year'),
    alt.Y('Firearm murders:Q', 
          title='Number of Firearm Murders',
          scale=alt.Scale(domain=[0, 1000])) # Start at 0 for honesty
)

line = base.mark_line(point=True, color='steelblue').properties(
    title='Trend of Firearm Murders in Florida (1991-2015)'
)

# 2. Add a vertical line for the law enactment (2005)
rule = alt.Chart(pd.DataFrame({'Year': [2005]})).mark_rule(
    color='red', 
    strokeDash=[5, 5],
    size=2
).encode(x='Year:O')

# 3. Add text label to explain the red line
text = alt.Chart(pd.DataFrame({
    'Year': [2005], 
    'murders': [900], 
    'text': ['Law Enacted (2005)']
})).mark_text(
    align='left',
    dx=10,
    color='red',
    fontWeight='bold'
).encode(
    x='Year:O',
    y=alt.Y('murders:Q'),
    text='text'
)

# Combine the layers
improved_guns_plot = (line + rule + text).properties(
    width=600,
    height=400
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=16,
    anchor='start'
)

improved_guns_plot
```

YOUR ANSWER HERE

The mayors claim that their approval ratings are "astronomical" is a textbook example of how to lie with statistics. By using a density plot to create smooth, professional-looking curves, they have managed to hide the fact that the entire dataset for each state consists of only three individual responses. In reality, you cannot generate a meaningful "distribution" from such a tiny sample size (); the software is simply filling in the gaps with mathematical guesswork to make the data look more robust than it actually is. Furthermore, the claim conveniently brushes aside the fact that one-third of the sampled voters in their own state gave them a rock-bottom score of . Calling this difference "statistically significant" is completely baseless, as the sample is far too small to represent the general public or to draw any valid comparison between the two states.

### Question 4.2
rubric={accuracy:2}

<h4>Python</h4>

```python

# Load the data
df_state = pd.read_csv('./data//state-comparison.csv')

# Recreate the misleading density plot
q4_2_plot = alt.Chart(df_state).transform_density(
    'the value',
    as_=['the value', 'density'],
    groupby=['where'],
    bandwidth=1000,    # Controls the "smoothness" of the hump
    extent=[0, 10000]   # Ensures the curve spans the whole axis
).mark_area(opacity=0.5).encode(
    alt.X('the value:Q', 
          title='Approval Rating (0 - 10,000)',
          scale=alt.Scale(domain=[0, 10000])),
    alt.Y('density:Q', title='Density'),
    alt.Color('where:N', 
              title='Location', 
              scale=alt.Scale(range=['#4c78a8', '#e15759'])) # Specific blue and red/orange
).properties(
    title='My Approval Ratings are Astronomical!',
    width=500,
    height=300
).configure_title(
    fontSize=18,
    anchor='start'
).configure_view(
    strokeWidth=0
)

q4_2_plot
```

### Question 4.3 (Optional)
rubric={viz:1}

<h4>Python</h4>
<p>Improve my Altair plot. Chose the best type of plot to accurately reveal what the data looks like. Design the plot to be an effective plot for communication purposes, this includes proper figure size, axis/figure titles, font sizes, colors, etc.</p>

```python
# YOUR ANSWER HERE
# Load the data
df_state = pd.read_csv('./data//state-comparison.csv')

# 1. Create a strip plot to show every single data point
strips = alt.Chart(df_state).mark_tick(
    size=30, 
    thickness=4
).encode(
    alt.X('the value:Q', 
          title='Approval Rating (0 - 10,000)',
          scale=alt.Scale(domain=[0, 10500])), # Give a little extra room for labels
    alt.Y('where:N', title='State/Location'),
    alt.Color('where:N', legend=None)
)

# 2. Add text labels above the ticks to show the exact values
text = strips.mark_text(
    align='center',
    baseline='bottom',
    dy=-15,
    fontWeight='bold'
).encode(
    text='the value:Q'
)

# 3. Combine and add an informative title
improved_state_plot = (strips + text).properties(
    title={
        "text": ["Comparison of Individual Approval Ratings"],
        "subtitle": ["Note: Extremely small sample size (n=3 per state) reveals high polarization, not consistent support."],
        "fontSize": 16,
        "subtitleFontSize": 12,
        "anchor": "start"
    },
    width=500,
    height=150
).configure_view(
    strokeWidth=0
)

improved_state_plot
```
